```python
from PythonDataProcessing import DataRetrieval as DR
```

```python
data = DR.get_stock_data('AAPL',interval='1d',period='2y')
```

```python
display(data)
```

```python
data = DR.add_EMA(data)
data = DR.add_SMA(data)
display(data)
```

```python
data["SMA"] = data["SMA"].shift(periods=1)
display(data)
```

```python
data =  data.iloc[1:, :]
display(data)
```

```python
labels = data["SMA"].values
labels = [(labels[i]-labels[i-1]) for i in range(1,len(labels))]
labels = [1 if i > 0 else 0 for i in labels]
labels = [99]+ labels
data["labels"] = labels
display(data)
```

```python
data = data.iloc[2:, :]
display(data)
```

```python
train_size = int(len(data) *0.8)
input_vals = ["High","Low","Close","Volume","EMA"]
label_vals = ["labels"]
print(train_size)
```

```python
from pandas.core.frame import DataFrame
from sklearn.preprocessing import MinMaxScaler

def fit_scalers(data_source: DataFrame, train_size, input_vals, label_vals):
    data_scaler = MinMaxScaler()
    labl_scaler = MinMaxScaler()

    data_scaler.fit(data_source[input_vals].iloc[:train_size])
    labl_scaler.fit(data_source[label_vals].iloc[:train_size])

    return data_scaler, labl_scaler

def normalize_stock_data(data_source, data_scaler, labl_scaler, input_vals, label_vals):
    scaled_data = data_scaler.transform(data_source[input_vals])
    scaled_lbls = labl_scaler.transform(data_source[label_vals])
    return scaled_data, scaled_lbls
```

```python
data_scaler, label_scaler = fit_scalers(data,train_size,input_vals,label_vals)
scaled_data, scaled_labels = normalize_stock_data(data, data_scaler, label_scaler, input_vals, label_vals)
display(scaled_data)
```

```python
import numpy as np
x = np.array(scaled_data)
y = np.ravel(scaled_labels)

train_x = x[:train_size]
train_y = y[:train_size]
test_x = x[train_size:]
test_y = y[train_size:]

print(len(train_x))
print(len(train_y))
print(len(test_x))
print(len(test_y))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(8,8,8), activation='relu', solver='adam', max_iter=500)
mlp.fit(train_x,train_y)
```

```python
pred_y = mlp.predict(train_x)
```

```python
from PythonDataProcessing import Metrics as MET
rndm_acc, acc = MET.get_acc(train_y,pred_y)
print(rndm_acc)
print(acc)
len(test_y)
```
